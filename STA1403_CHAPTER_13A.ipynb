### **_13.2 A Simple Case of Bayesian Analysis for Population Proportion_**

We start our discussion with a simple illustrative example. Suppose that we are interested in finding the five-year survival rate (i.e., population proportion of survival) among breast cancer patients. We denote this unknown population proportion $ \mu $ . For simplicity, we assume that the survival rate is either 0.75 or 0.85. Without any data, we think that both of these values are equally probable; that is, $P(\mu = 0.75) = P(\mu = 0.85) = 0.5 $. Alternatively, we can write this as follows:
$$
\frac {P (\mu = 0 . 8 5)}{P (\mu = 0 . 7 5)} = 1.
$$
Now suppose that we take a random sample of $n$ = 20 breast cancer patients from the population. We use $Y$ to denote the number of survivals out of $20$. From Chap. 5 we know that $Y$ has a Binomial($n$, $\mu$) distribution (assuming that the patients are selected independently and they all have the same probability of survival). Therefore, if $\mu = 0.75$, the distribution of $Y$ is $Binomial(20, 0.75)$. If $\mu = 0.85$ on the other hand, the distribution of $Y$ is $Binomial(20, 0.85)$.

For our sample, we find that $18$ of patients are still alive after five years: $Y = 18$. Our point estimate for the survival rate $\mu$ is therefore $p = 18/20 = 0.9$, where $p$ is the sample proportion. Let us see how this information changes our mind about the value of the population proportion, $\mu$. For this, we use Bayes' theorem. Recall that Bayes' formula for two events $ E_{1}$ and $ E_{2}$ is
$$
P \left(E_{2} \mid E_{1}\right) = \frac{P \left(E_{1} \mid E_{2}\right) P \left(E_{2}\right)}{P \left(E_ {1}\right)}.
$$

For our example, we can write Bayes' formula in terms of $\mu$ and $Y$, so $E_{1}$ corresponds to the event that $Y=18$, and $E_{2}$ corresponds to the event that $\mu=0.85$:
$$
P (\mu = 0.85 | Y = 18) = \frac {P (Y = 18 | \mu = 0.85) P(\mu = 0.85)}{P(Y = 18)}.
$$
Here, $P(\mu = 0.85|Y = 18)$ is the probability that the true value of the survival rate is 0.85 given the information that 18 people have survived (out of 20), $P(Y = 18|\mu = 0.85) $ is the probability that 18 people survive assuming that the true survival rate is $0.85$, $P(\mu = 0.85)$ is the probability that the survival rate is in fact 0.85 (we assumed this probability is $0.5$), and $ P(Y = 18)$ is the probability that $18$ people survive regardless of the what the probability of survival is ($0.85$ or $0.75$).

As mentioned above, $Y$ has Binomial($20$, $0.85$) distribution given that $\mu = 0.85$. Using Python, we can find the probability of $18$ survivals assuming that $\mu = 0.85$.

Click Distributions $ \rightarrow $ Discrete distributions $ \rightarrow $ Binomial distribution $ \rightarrow $ Binomial probabilities. Then, set Binomial trials to 20 and Probability of success to 0.85. Python provides the probabilities for all possible values of Y. The probability for $18$ survivals assuming that $\mu = 0.85$ is $P(Y = 18|\mu = 0.85) = 0.23$.

To find $P(Y = 18)$, we use the law of total probability (Eq. 4.7):
$$
\begin{array}{l} P (Y = 18)= P(Y = 18 | \mu = 0.85) P (\mu = 0.85) \ + P (Y = 18 | \mu = 0.75) P (\mu = 0.75). \\
\end{array}
$$
Here, $P(Y = 18|\mu = 0.75)$ is the probability of $18$ survivals assuming that the true value of the survival rate is $ \mu = 0.75 $ . We find this probability using Python by following the above steps, but this time we set Probability of success to 0.75. By doing so, we find $ P(Y = 18|\mu = 0.75) = 0.07$. Therefore,
$$
\begin{array}{l} P(Y = 18) = 0.23 \times 0.5 + 0.07 \times 0.5 \\
= 0.15. \\
\end{array}
$$
Now we have all the information we need to find $P(\mu = 0.85|Y = 18)$:
$$
\begin{array}{l} P(\mu = 0.85|Y = 18) = \frac{P (Y = 18 | \mu = 0.85) P(\mu = 0.85)}{P(Y = 18)} \\
= \frac {0.23 \times 0.5}{0.15} \\
= 0.76. \\
\end{array}
$$
At the beginning (before observing any data), we believed that $ \mu = 0.85 $ with probability of 0.5. Knowing that 18 out of 20 people have survived, we increase this probability to 0.76.

By following similar steps, we find that $ P(\mu = 0.75|Y = 18) = 0.24 $ . Given the observed data, we have reduced the probability of $ \mu = 0.75 $ from 0.5 to 0.24. Therefore, while we gave equal probabilities to both values 0.75 and 0.85 at the beginning, based on the new empirical evidence we observed, we increased the probability of $ \mu = 0.85 $ and decreased the probability of $ \mu = 0.75 $ . This is intuitive, of course, because our point estimate for the survival rate is 0.9 (18 out of 20), which is closer to 0.85 than 0.75. We can use these updated probabilities and write
$$
\frac {P (\mu = 0.85 | Y = 18)}{P (\mu = 0 .75 | Y = 18)} = \frac {0.76}{0.24} = 3.2.
$$
Therefore, given the observed data, the value $0.85$ is $3.2$ times more likely than $0.75$.

## Example 1: Bayesian Posterior Calculator (Binomial Likelihood)

The Python code in the cell below can be used to solve this simple case discussed above.

````python
# @title Example 1: Bayesian Posterior Calculator (Binomial Likelihood)

import scipy.stats as stats
import pandas as pd
from itertools import combinations

# ── INPUT SECTION ──────────────────────────────────────────────────────
n     = 20          # Total trials (e.g., patients treated)
y_obs = 18          # Observed successes (e.g., survivors)

# mu, prior pairs — priors must sum to 1.0
hypotheses = {
    0.85: 0.5,      # mu : prior probability P(mu)
    0.75: 0.5,
    # 0.90: 0.25,   # ← uncomment/add rows to explore more hypotheses
}
# ──────────────────────────────────────────────────────────────────────

def bayes_binomial(n, y_obs, hypotheses):
    total_prior = sum(hypotheses.values())
    if abs(total_prior - 1.0) > 1e-9:
        raise ValueError(f"Priors must sum to 1.0 (currently sum to {total_prior:.4f})")

    rows = []
    for mu, prior in hypotheses.items():
        likelihood = stats.binom.pmf(y_obs, n, mu)
        joint      = likelihood * prior
        rows.append({"mu": mu, "Prior": prior,
                     "Likelihood P(Y|mu)": likelihood, "Joint": joint})

    df = pd.DataFrame(rows)
    marginal = df["Joint"].sum()
    df["Posterior P(mu|Y)"] = df["Joint"] / marginal
    return df, marginal


def print_odds_ratios(results):
    """Print posterior odds ratios for every pair of hypotheses."""
    mu_vals = results["mu"].tolist()
    posts   = results.set_index("mu")["Posterior P(mu|Y)"]

    if len(mu_vals) < 2:
        return

    print("\nPosterior Odds Ratios")
    print("-" * 46)
    pairs = list(combinations(mu_vals, 2))
    for (a, b) in pairs:
        # Always put the higher-posterior mu in the numerator
        if posts[a] >= posts[b]:
            num, den = a, b
        else:
            num, den = b, a
        ratio = posts[num] / posts[den]
        print(f"  P(μ={num:.2f}|Y={y_obs}) / P(μ={den:.2f}|Y={y_obs})"
              f"  =  {posts[num]:.4f} / {posts[den]:.4f}"
              f"  =  {ratio:.2f}")
        print(f"  → μ={num:.2f} is {ratio:.2f}x more probable than μ={den:.2f}")
        if len(pairs) > 1:
            print()


# ── Run ────────────────────────────────────────────────────────────────
results, p_y = bayes_binomial(n, y_obs, hypotheses)

point_est = y_obs / n
print(f"\nBayesian Update  |  n={n},  Y={y_obs},  "
      f"point estimate μ̂ = {y_obs}/{n} = {point_est:.4f}\n")

# Full Bayes table
print(f"{'mu':>6}  {'Prior':>8}  {'P(Y|mu)':>12}  {'Joint':>10}  {'Posterior':>12}")
print("-" * 56)
for _, row in results.iterrows():
    print(f"{row['mu']:>6.2f}  {row['Prior']:>8.4f}  "
          f"{row['Likelihood P(Y|mu)']:>12.6f}  "
          f"{row['Joint']:>10.6f}  "
          f"{row['Posterior P(mu|Y)']:>12.6f}")
print("-" * 56)
print(f"{'Marginal  P(Y=' + str(y_obs) + ')':>40}  {p_y:>10.6f}")

# Best hypothesis
best = results.loc[results["Posterior P(mu|Y)"].idxmax()]
print(f"\n→ Most probable:  μ = {best['mu']:.2f}  "
      f"(posterior = {best['Posterior P(mu|Y)']:.4f},  "
      f"closest hypothesis to point estimate {point_est:.2f})")

# Odds ratios
print_odds_ratios(results)
````

## **13.3 Prior and Posterior Probabilities**

The above example illustrates a simple application of Bayesian inference. For the population proportion $\mu$ , we refer to $P(\mu = 0.75)$ and $P(\mu = 0.85)$ as **prior probabilities**. These are probabilities we assign to possible values of $\mu$ before observing any data. Note that in practice, we might obtain these probabilities from previous studies. For example, two other research groups might have conducted similar studies in the past; one group estimated $\mu$ to be $0.75$, and the other group estimated it to be $0.85$, and we do not have any reason to prefer one estimate over the other. In this case, we want to conduct a new study, collect new empirical evidence, and estimate $\mu$ , but we want to take the available information regarding the value of $\mu$ into account.

We refer to $P(Y = 18|\mu = 0.85)$ as **likelihood**, i.e., how likely it is to see this specific data ($18$ survivals out of $20$) if $\mu$ is in fact $0.85$. We can express the probability of the specific data we have observed (i.e., $18$ survivals out of $20$) as a function of different values of $\mu$. We refer to this function as the **likelihood function**. For the above example, the likelihood function is

$$P(Y = 18 \mid \mu) = \begin{cases} 0.07, & \mu = 0.75 \\ 0.23, & \mu = 0.85 \end{cases}$$

We refer to the updated probability of $\mu$ after we observe the data as the **posterior probability** of $\mu$. The posterior probabilities in the above example are $P(\mu = 0.75|Y = 18) = 0.24$ and $P(\mu = 0.85|Y = 18) = 0.76$, which are obtained after we observed $18$ survivors among $20$ patients. As mentioned above, we can use these posterior probabilities to write
$$
\frac {P (\mu = 0.85 | Y = 18)}{P (\mu = 0 .75 | Y = 18)} = \frac {0.76}{0.24} = 3.2.
$$

This is known as the **posterior odds** (here, we find the odds of 0.85 over 0.75). Of course, based on what we discussed above, we can find the posterior odds as follows:
$$
\begin{array}{l}
\frac {P (\mu = 0.85 | Y = 18)}{P (\mu = 0.75 | Y = 18)} = \frac {P(Y = 18 | \mu = 0.85) P(\mu = 0 .85) //
P (Y = 18)}{P (Y = 18 |\mu = 0.75) P(\mu = 0.75) //
P (Y = 18)} \\
= \frac {P(Y = 18 | \mu = 0.85) P (\mu = 0.85)}{P (Y = 18 | \mu = 0.75) P (\mu = 0.75)}. \\
\end{array}
$$
By canceling out the term $P(Y = 18)$ from the numerator and denominator, we have
$$
\frac {P (\mu = 0.85 | Y = 18)}{P (\mu = 0.75 | Y = 18)} = \frac {P (Y =18|\mu = 0.85)}{P (Y = 18|\mu = 0.75)} \times \frac {P(\mu = 0.85)}{P (\mu = 0.75)}.
$$

The term $ P(\mu = 0.85)/P(\mu = 0.75) $ on the right-hand side of the above equation is called **prior odds** (here, we find the odds of 0.85 over 0.75). In our example, the prior odds is 1. The posterior odds is obtained by multiplying the prior odds by the following term:

$$
\frac {P (Y =18 | \mu = 0.85)}{P (Y = 18 | \mu = 0.75)} = \frac {0.23}{0.07}.
$$

This term is in fact the ratio of two possible values for the likelihood function and is known as the **likelihood ratio**.

----------------

>The posterior odds is the product of the prior odds by the likelihood ratio.

----------------

## Example 2: Calculate Prior, Likelihood, Posterior & Odds

The code in the cell below can be used to calculate the prior, likelihood, posterior and odds.

````python
# @title Example 2: Calculate Prior, Likelihood, Posterior & Odds

import scipy.stats as stats
import pandas as pd
from itertools import combinations

# ── INPUT SECTION ───────────────────────────────────────────────────────
n     = 20          # Total trials (e.g., patients treated)
y_obs = 18          # Observed successes (e.g., survivors)

# Prior probabilities for each hypothesis — must sum to 1.0
hypotheses = {
    0.85: 0.5,      # mu : P(mu)  ← prior probability
    0.75: 0.5,
}

# Which pair of mu values to use for odds calculations?
mu_A = 0.85         # numerator   (the "favored" hypothesis)
mu_B = 0.75         # denominator (the "alternative" hypothesis)
# ──────────────────────────────────────────────────────────────────────


def bayes_binomial(n, y_obs, hypotheses):
    total_prior = sum(hypotheses.values())
    if abs(total_prior - 1.0) > 1e-9:
        raise ValueError(
            f"Priors must sum to 1.0 (currently {total_prior:.4f}). "
            "See Section 13.3 — prior probabilities represent your full "
            "belief over all possible values of mu."
        )
    rows = []
    for mu, prior in hypotheses.items():
        likelihood = stats.binom.pmf(y_obs, n, mu)
        joint      = likelihood * prior
        rows.append({"mu": mu, "Prior P(mu)": prior,
                     "Likelihood P(Y|mu)": likelihood, "Joint": joint})
    df       = pd.DataFrame(rows).sort_values("mu", ascending=False)
    marginal = df["Joint"].sum()                    # P(Y = y_obs)
    df["Posterior P(mu|Y)"] = df["Joint"] / marginal
    return df, marginal


def print_likelihood_function(results, y_obs, n):
    """Display the likelihood function P(Y=y_obs | mu) for all hypotheses."""
    print(f"  Likelihood function  P(Y={y_obs} | mu)  [n={n}]")
    print("  " + "-" * 30)
    for _, row in results.iterrows():
        print(f"    mu = {row['mu']:.2f}  →  P(Y={y_obs}|mu) = {row['Likelihood P(Y|mu)']:.4f}")


def print_odds_analysis(results, mu_A, mu_B, y_obs):
    """
    Decompose posterior odds into prior odds × likelihood ratio.
    Mirrors the textbook formula:
        P(mu_A|Y) / P(mu_B|Y)  =  [P(Y|mu_A)/P(Y|mu_B)]  ×  [P(mu_A)/P(mu_B)]
    """
    idx = results.set_index("mu")
    if mu_A not in idx.index or mu_B not in idx.index:
        print(f"\n  ⚠  mu_A={mu_A} or mu_B={mu_B} not found in hypotheses dict.")
        return

    prior_A      = idx.loc[mu_A, "Prior P(mu)"]
    prior_B      = idx.loc[mu_B, "Prior P(mu)"]
    like_A       = idx.loc[mu_A, "Likelihood P(Y|mu)"]
    like_B       = idx.loc[mu_B, "Likelihood P(Y|mu)"]
    post_A       = idx.loc[mu_A, "Posterior P(mu|Y)"]
    post_B       = idx.loc[mu_B, "Posterior P(mu|Y)"]

    prior_odds   = prior_A / prior_B
    like_ratio   = like_A  / like_B
    post_odds    = post_A  / post_B

    print(f"\n{'─'*56}")
    print(f"  Odds Analysis:  μ={mu_A:.2f}  vs  μ={mu_B:.2f}")
    print(f"{'─'*56}")
    print(f"  Prior odds       P(μ={mu_A:.2f}) / P(μ={mu_B:.2f})")
    print(f"                 = {prior_A:.4f} / {prior_B:.4f}  =  {prior_odds:.4f}")

    print(f"\n  Likelihood ratio  P(Y={y_obs}|μ={mu_A:.2f}) / P(Y={y_obs}|μ={mu_B:.2f})")
    print(f"                 = {like_A:.4f} / {like_B:.4f}  =  {like_ratio:.4f}")

    print(f"\n  Posterior odds   = Prior odds × Likelihood ratio")
    print(f"                 = {prior_odds:.4f} × {like_ratio:.4f}  =  {post_odds:.4f}")

    print(f"\n  Verification from posteriors directly:")
    print(f"    P(μ={mu_A:.2f}|Y={y_obs}) / P(μ={mu_B:.2f}|Y={y_obs})")
    print(f"  = {post_A:.4f} / {post_B:.4f}  =  {post_A/post_B:.4f}  ✓")
    print(f"\n  → μ={mu_A:.2f} is {post_odds:.2f}x more probable than μ={mu_B:.2f} "
          f"given Y={y_obs}")


# ── Run ────────────────────────────────────────────────────────────────
results, p_y = bayes_binomial(n, y_obs, hypotheses)
point_est    = y_obs / n

print(f"\n{'═'*56}")
print(f"  Bayesian Inference  |  n={n},  Y={y_obs}")
print(f"  Point estimate  μ̂ = {y_obs}/{n} = {point_est:.4f}")
print(f"{'═'*56}")

# 1. Likelihood function
print("\n① Likelihood Function (prior to updating)\n")
print_likelihood_function(results, y_obs, n)

# 2. Full Bayes table
print(f"\n② Bayes Table\n")
print(f"  {'mu':>5}  {'Prior':>10}  {'Likelihood':>12}  {'Joint':>10}  {'Posterior':>12}")
print("  " + "-" * 56)
for _, row in results.iterrows():
    print(f"  {row['mu']:>5.2f}  {row['Prior P(mu)']:>10.4f}  "
          f"{row['Likelihood P(Y|mu)']:>12.6f}  "
          f"{row['Joint']:>10.6f}  "
          f"{row['Posterior P(mu|Y)']:>12.6f}")
print("  " + "-" * 56)
print(f"  {'Marginal  P(Y=' + str(y_obs) + ')':>42}  {p_y:>10.6f}")

# 3. Odds decomposition
print(f"\n③ Posterior Odds Decomposition")
print_odds_analysis(results, mu_A, mu_B, y_obs)

# 4. Most probable hypothesis
best = results.loc[results["Posterior P(mu|Y)"].idxmax()]
print(f"\n{'═'*56}")
print(f"  Most probable:  μ = {best['mu']:.2f}  "
      f"(posterior = {best['Posterior P(mu|Y)']:.4f})")
print(f"  (Point estimate {point_est:.2f} is closest to μ={best['mu']:.2f})")
print(f"{'═'*56}\n")
````
